```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for plots
%matplotlib inline
sns.set_style('white')
```

```python
import glob
import time
import datetime

path ='ydata-labeled-time-series-anomalies-v1_0/A3Benchmark/'
allFiles = glob.glob(path + "/*.csv")
raw_data = pd.DataFrame()
list_dfs = []

def date_parser(string_list):
    #try:
    #    int_val = int(string_list)
    #except ValueError:
    #    print('failure with timestamp', string_list)
    return [time.ctime(float(x)) for x in string_list]

#def date_parser(string_list):
#    return datetime.datetime.fromtimestamp(int(string_list))
    
i = 0
for f in allFiles:
    #df = pd.read_csv(f,sep=',',skiprows=1,header=None, index_col=0,date_parser=date_parser) #setting timestamp as index
    df = pd.read_csv(f,sep=',',skiprows=1,header=None, index_col=None) #setting timestamp as index
    if df.shape[1] == 9: # skipping the "all" file
        list_dfs.append(df)
        i+= 1

raw_data = pd.concat(list_dfs)
raw_data.columns = ['timestamp', 'value', 'anomaly', 'changepoint', 'trend', 'noise', 'seasonality1', 'seasonality2', 'seasonality3']
pd.to_numeric(raw_data['anomaly'])
print('data dimensions', raw_data.shape)
print('# of files imported:', i)
```

```python
# List of column names containing all the features for regression.
features = raw_data.columns.difference(['anomaly'])
features =  ['value', 'trend', 'noise']

# The name of the column for the output varaible.
target = 'anomaly'
#print('Features: {}'.format(features.values))
print(type(features))
```

```python
from sklearn.cross_validation import train_test_split

x = raw_data[features]
y = raw_data[target]

# Divide the data into a training and a test set.
random_state = 0  # Fixed so that everybody has got the same split
test_set_fraction = 0.2
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_set_fraction, 
                                                    random_state=random_state)

print('Size of training set: {}'.format(len(x_train)))
print('Size of test set: {}'.format(len(x_test)))
```

```python
from sklearn import preprocessing

features_data = raw_data[features]

scaler = preprocessing.StandardScaler()
scaler.fit(features_data)
x_test = scaler.transform(x_test)
x_train = scaler.transform(x_train)
```

```python
#from sklearn import svm

# Train the classifier.
#method = svm.SVC(kernel='rbf', C=100.)
#method.fit(x_train, y_train)
```

```python
from sklearn import ensemble
from sklearn import *

method = ensemble.RandomForestClassifier()
method.fit(x_train, y_train)
method.score(x_test,y_test)
```

```python
# Evaluate on test data.
print('Accuracy =', method.score(x_test, y_test))
```

```python
from sklearn.metrics import classification_report

print(classification_report(x_test, y_test))
```

```python
from sklearn.metrics import roc_curve

y_score = method.decision_function(x_test)
fpr, tpr, uh = roc_curve(y_test, y_score)

plt.rcParams['figure.figsize'] = (6.0, 6.0)
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlabel('False negatives')
plt.ylabel('False positives')
```
